```python

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
blood_sugar = np.concatenate([
    np.random.normal(90, 10, 120),
    np.random.normal(150, 15, 80)
])

sns.kdeplot(blood_sugar, fill=True, color='blue', alpha=0.4)
plt.title("Density Plot of Blood Sugar Levels (200 Patients)")
plt.xlabel("Blood Sugar Level (mg/dL)")
plt.ylabel("Density")


plt.show()
```

```python

import pandas as pd
import plotly.express as px

data = {
    'State': ['Maharashtra', 'Kerala', 'Delhi', 'Karnataka', 'Tamil Nadu'],
    'Cases': [320, 250, 150, 180, 200],
    'Latitude': [19.7515, 10.8505, 28.7041, 15.3173, 11.1271],
    'Longitude': [75.7139, 76.2711, 77.1025, 75.7139, 78.6569]
}

df = pd.DataFrame(data)

fig = px.scatter_geo(
    df,
    lat='Latitude',
    lon='Longitude',
    color='Cases',
    size='Cases',
    hover_name='State',
    projection='natural earth',
    color_continuous_scale='Reds',
    title='COVID-19 Cases Across Indian States (2020)',
)

fig.update_geos(scope='asia', center={'lat': 22, 'lon': 80}, projection_scale=4)
fig.update_layout(margin={"r":0,"t":40,"l":0,"b":0})
fig.show()

```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
study_hours = np.random.normal(5, 1.5, 50)
sleep_hours = np.random.normal(7, 1, 50)
exam_score = 10 * study_hours - 2 * sleep_hours + np.random.normal(0, 5, 50)

df = pd.DataFrame({
    'Study Hours': study_hours,
    'Sleep Hours': sleep_hours,
    'Exam Score': exam_score
})

sns.pairplot(df, diag_kind='kde', palette='coolwarm')
plt.suptitle("Pair Plot  Study Hours, Sleep Hours, Exam Scores", y=1.02)
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(10)
df = pd.DataFrame({
    'Ratings': np.random.normal(7, 1, 100),
    'Budget': np.random.normal(50, 15, 100),
    'BoxOffice': np.random.normal(120, 30, 100),
    'Genre': np.random.choice(['Action', 'Comedy', 'Drama', 'Horror'], 100)
})

plt.figure(figsize=(6,4))
sns.histplot(df['Ratings'], kde=True, color='skyblue')
plt.title("Histogram  Movie Ratings")
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x='Genre', y='BoxOffice', data=df, palette='viridis')
plt.title("Box Plot  Box Office by Genre")
plt.show()

plt.figure(figsize=(6,4))
sns.kdeplot(df['Budget'], fill=True, color='orange')
plt.title("Density Plot  Budget Distribution")
plt.show()

sns.pairplot(df[['Ratings','Budget','BoxOffice']], diag_kind='kde', palette='coolwarm')
plt.suptitle("Pair Plot  Ratings, Budget, BoxOffice", y=1.02)
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(20)
df = pd.DataFrame({
    'Math': np.random.randint(50, 100, 100),
    'Science': np.random.randint(50, 100, 100),
    'English': np.random.randint(50, 100, 100),
    'Computer': np.random.randint(50, 100, 100)
})

corr = df.corr()
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Heatmap  Student Performance Correlation")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Category': ['Electronics', 'Clothing', 'Furniture', 'Groceries'],
    'Q1': [50, 30, 20, 40],
    'Q2': [60, 40, 25, 45],
    'Q3': [70, 35, 30, 50],
    'Q4': [120, 50, 40, 60]
}

df = pd.DataFrame(data)
df.set_index('Category', inplace=True)

plt.figure(figsize=(6,4))
sns.heatmap(df, annot=True, cmap='YlOrRd')
plt.title("Heatmap  Quarterly Sales by Category ( Lakhs)")
plt.show()

```
